```python
import random

def random_list(start, end, size):
    result = []
    
    for i in range(size):
        result.append(random.uniform(start, end))
        
    return result

X = random_list(36, 38, 3000)
Y = random_list(127, 127.3, 3000)

random_location = pd.DataFrame({'latitude': X, 'longtitude' : Y})
random_location.to_csv('./data/random_location.csv', header = True, index = False)

# data = {'X': X, 'Y' : Y}
```

```python
import pandas as pd
data = pd.read_csv("./data/random_location.csv")
print(data.shape)
data.head()
```

3000개의 데이터를 지도에 핀으로 표시하게 되면 지도가 정갈하지 X. 따라서 folium 플러그인 중 MakerCluster을 이용하여 클러스터 형성.

```python
from folium.plugins import MarkerCluster
```

```python
def plot(df):
    locations = list(zip(df.latitude, df.longtitude))
    
    icons = [folium.Icon(icon="car", prefix="fa", color="red") for i in range(len(locations))]
                         
    cluster = MarkerCluster(locations = locations, icons = icons)
    cluster.add_to(m)
```

```python
plot(data)
m.save('./data/random_data.html')
```
