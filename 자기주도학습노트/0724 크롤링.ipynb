```python
from bs4 import BeautifulSoup as bs
from pprint import pprint
import requests

html = requests.get('https://search.naver.com/search.naver?query=날씨')
# pprint(html.text)
```

```python
soup = bs(html.text,'html.parser')
# pprint(soup)
```

```python
data1 = soup.find('div',{'class':'detail_box'})   # 영역 추출

# 여러 개의 동일한 div/class값이 있을 경우, 가장 첫 번째 있는 것만 추출됨.
```

```python
data2 = data1.findAll('dd')
# pprint(data2)
```

```python
fine_dust = data2[0].find('span',{'class':'num'})
print(fine_dust)
```

```python
fine_dust = data2[0].find('span',{'class':'num'}).text   # 텍스트만 불러 오는 것.
print(fine_dust)
```

```python
# 웹 페이지를 열고 소스코드 읽어오는 작업.
html = requests.get("https://comic.naver.com/webtoon/weekday")
soup = bs(html.text, 'html.parser')
html.close()
```

```python
# 월요웹툰 영역 html 추출하기.
data1 = soup.find('div',{'class':'col_inner'})
pprint(data1)
```

```python
# 제목포함 영역 html 추출하기.
data2 = data1.findAll('a', {'class':'title'})
# pprint(data2)

# 텍스트만 추출
# title_list = []
# for i in data2:
#     title_list.append(i.text)

# for 문 list appending 한줄로 하는 방법
title_list = [i.text for i in data2]
    
pprint(title_list)
```

```python
# 모든 요일 웹툰 제목 가져오기
# 현재 작업
# [월요 웹툰 영역 find] > [해당 영역 제목 findAll] > [for문으로 text 추출]
```

```python
# find => findAll 로 모든 요일 영역 가져오기.

data1 = soup.findAll('div', {'class':'col_inner'})
```
