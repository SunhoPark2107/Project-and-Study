```python
from bs4 import BeautifulSoup as bs
from pprint import pprint
import requests
import pandas as pd
```

```python
html = requests.get('http://www.egalchi.com/shop/list.php?ca_id=1060')
soup = bs(html.text,'html.parser')

data1 = soup.find('div',{'class':'Product link2'})

data_title = data1.findAll('span', {'class' : 'Title'})
data_price = data1.findAll('span', {'class' : 'Price'})

title_list = [i.text for i in data_title]
price_list = [i.text for i in data_price]

products = {'상품명':title_list,
           '가격':price_list}

df = pd.DataFrame(products, columns = ['상품명','가격'])

df
```

```python
category_list = []
title_list = []
price_list = []

for a in range(1, 10):
    url = str('http://www.egalchi.com/shop/list.php?ca_id=10'+str(a)+'0')

    
    html = requests.get(url)
    soup = bs(html.text,'html.parser')
    
    data1 = soup.find('div',{'class':'Product link2'})
    
    data_title = data1.findAll('span', {'class' : 'Title'})
    data_price = data1.findAll('span', {'class' : 'Price'})
    
    title_list_a = [i.text for i in data_title]
    price_list_a = [i.text for i in data_price]
    
    
    data_index = soup.find('div',{'class':'Submenu'})
    data_index2 = data_index.findAll('a')
    
    
    # rfind 함수 => a.rfind() a 뭉치에서 괄호 안의 내용이 있으면 해당 내용을 반환하고, 없다면 -1을 반환함.(array형식으로 토하는듯.)
    for Text in data_index2 :
        if str(Text).rfind('10'+str(a)+'0') != -1:
            category = Text.text
    
    
    
    for i in range(len(title_list_a)):
        category_list.append(category)
    title_list.extend(title_list_a)
    price_list.extend(price_list_a)


print(category_list)
print(title_list)
print(price_list)
```
